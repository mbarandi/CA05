**Question:** Given a movies data set, what are the 5 most similar movies to a movie query?

```python
#Importing necessary packages
import pandas as pd
import numpy as np
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
```

```python
#Importing the data
df = pd.read_csv('https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv')
df
```

**Building Our Own Recommender System**

```python
#Assigning our columns to x and y variables to begin analysis
x_label = df[['IMDB Rating', 'Biography', 'Drama','Thriller', 'Comedy', 'Crime', 'Mystery', 'History']]

#Setting a class column and assigning it to a new variable 'y'
y_label = df['Movie Name']
```

```python
#Defining the kNN model and fitting it on the feature and class data
knn_model = NearestNeighbors(n_neighbors=5).fit(x_label,y_label)
```

**Genre Information about the movie "The Post"**

**IMDB Rating =** 7.2

**Biography =** Yes

**Drama =** Yes

**Thriller =** No

**Comedy =** No

**Crime =** No

**Mystery =** No

**History =** Yes

```python
#List containing characteristics of The Post
the_post = [7.2, 1, 1, 0, 0, 0, 0, 1]
```

```python
#Using k-NN to find nearest data points to our query 
distances, indices = knn_model.kneighbors([the_post])
print(indices)
```

```python
#Extracting recommended indices from k-NN output
recommended_indices = indices[0]
```

```python
#Extracting recommended movie names from indices
recommended_movies = df.iloc[recommended_indices]['Movie Name'].values
```

```python
#Printing the recommended movies
print(recommended_movies)
```

**Recommendation**

```python
#Printing out our 5 recommended movies based on The Post
print("Five Movies Similar to 'The Post'")
for title in recommended_movies:
    print(title)
     
```
